```python
#elbow method to determine the optimal number of clusters
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_train_scaled)
    inertia.append(kmeans.inertia_)
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from matplotlib.patches import Ellipse

def plot_clusters():
    X, y = make_blobs(n_samples=300, centers=[(2,2), (6,6), (10,10)], cluster_std=[1.0, 1.5, 2.0], random_state=42)
    
    # Define cluster labels
    cluster_labels = ['Normal', 'Mild', 'Severe']
    colors = ['blue', 'orange', 'red']
    
    # Plot data points
    plt.figure(figsize=(8, 6))
    for i in range(3):
        plt.scatter(X[y == i, 0], X[y == i, 1], label=cluster_labels[i], color=colors[i], alpha=0.6)
        
        # Draw an ellipse around each cluster
        cov_matrix = np.cov(X[y == i].T)
        eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)
        angle = np.degrees(np.arctan2(*eigenvectors[:,0][::-1]))
        width, height = 2 * np.sqrt(eigenvalues)
        ellipse = Ellipse(xy=np.mean(X[y == i], axis=0), width=width*2, height=height*2, 
                          angle=angle, edgecolor=colors[i], facecolor='none', linewidth=2)
        plt.gca().add_patch(ellipse)
    
    # Customize plot
    plt.legend()
    plt.xlabel('Quality of Sleep')
    plt.ylabel('Stress Level')
    plt.title('Clustered Data Points with Ellipse Boundaries')
    plt.grid(True)
    plt.show()

# Call the function to plot clusters
plot_clusters()
```
